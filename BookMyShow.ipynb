```python
import random

class User:
    def __init__(self):
        self.email = ""
        self.mobile = ""
        self.password = ""
        self.otp = 0

    # Signup method
    def signup(self):
        # Email validation
        while True:
            self.email = input("Enter email (@gmail.com only): ")
            if self.email.endswith("@gmail.com"):
                break
            else:
                print("Invalid email! Email must end with @gmail.com")

        # Mobile number validation (10 digits)
        while True:
            self.mobile = input("Enter mobile number (10 digits): ")
            if self.mobile.isdigit() and len(self.mobile) == 10:
                break
            else:
                print("Invalid mobile number!")

        # Password validation
        while True:
            self.password = input(
                "Enter password (1 uppercase, 1 lowercase, 1 @): "
            )
            if self.validate_password(self.password):
                break
            else:
                print("Invalid password! Try again.")

    # Password validation logic
    def validate_password(self, password):
        has_upper = False
        has_lower = False
        has_at = False

        for char in password:
            if char.isupper():
                has_upper = True
            if char.islower():
                has_lower = True
            if char == "@":
                has_at = True

        return has_upper and has_lower and has_at

    # Generate and send OTP
    def send_otp(self):
        self.otp = random.randint(100000, 999999)
        print("OTP sent to email/mobile:", self.otp)

    # Verify OTP
    def verify_otp(self):
        entered_otp = int(input("Enter OTP: "))
        if entered_otp == self.otp:
            self.logged_in = True
            print("OTP verified")
            print("User logged in successfully.")

            print("====================================================================")
            print("            Welcome To Our Book My Show!!! Have a Great Day         ")
            print("====================================================================")

        else:
            print("Wrong OTP.")


object=User()
object.signup()
object.send_otp()
object.verify_otp()
```

```python
class Movie:
    def show_movies(self):
        print("Choose Your Mood")
        print("1. Romantic")
        print("2. Relaxed")
        print("3. Excited")
        print("4. Family")

        choice = input("Enter choice: ")

        if choice == "1":
            print("\nRomantic Movies:")
            movies = ["Love", "Romance", "Forever", "Heart Touch", "True Love"]
        elif choice == "2":
            print("\nRelaxed Movies:")
            movies = ["Comedy Night", "Smile Time", "Fun Day"]
        elif choice == "3":
            print("\nExcited Movies:")
            movies = ["Fast Action", "Thrill Ride", "Race Time"]
        elif choice == "4":
            print("\nFamily Movies:")
            movies = ["Family Bond", "Home Story", "Happy Family"]
        else:
            print("Invalid choice")
            return 

        for movie in movies:
            print("-", movie)
            
```

```python
obj=Movie()
obj.show_movies()
```
